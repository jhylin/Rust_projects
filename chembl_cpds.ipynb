The dataset shows a fully integrated dataframe with all data packed as strings for each compound in each row. Each variable is separated by semicolons. To read the dataframe properly, let's separate them and transform the dataframe to a more readable format.

```python
# Going back to polars documentation, use "sep" to set the delimiter of the file
# which in this case is semicolon
df = pl.read_csv("chembl_mols.csv", sep = ";")
# Show the first 10 rows of data
df.head(10)
```

```python
# List all column names to see what compound properties are stored in chEMBL database
df_col_list = list(df.columns)
df_col_list
```

As shown above, a separate dataframe was saved as df_col_list to show all the column names used, which were equivalent to the physicochemical properties of small molecules curated by ChEMBL database.

Selected definitions of the physicochemical properties of compounds were adapted from ChEMBL_31 schema documentation (available as "Release notes" on website) as shown below:

'Max Phase' = Maximum phase of development reached for the compound (4 = approved). Null where max phase has not yet been assigned.

'AlogP' = Calculated partition coefficient

'HBA' = Number of hydrogen bond acceptors

'HBD' = Number of hydrogen bond donors

'#RO5 Violations' = Number of violations of Lipinski's rule-of-five, using HBA and HBD definitions

'Passes Ro3' = Indicates whether the compound passes the rule-of-three (mw < 300, logP < 3 etc)

'QED Weighted' = Weighted quantitative estimate of drug likeness (as defined by Bickerton et al., Nature Chem 2012)

'CX Acidic pKa' = 	The most acidic pKa calculated using ChemAxon v17.29.0

'CX Basic pKa' = The most basic pKa calculated using ChemAxon v17.29.0
 
'CX LogP' = The calculated octanol/water partition coefficient using ChemAxon v17.29.0
 
'CX LogD' = The calculated octanol/water distribution coefficient at pH7.4 using ChemAxon v17.29.0

'Inchi Key' = the IUPAC international chemical identifier key

```python
# Checking if there were any names recorded in the "Name" column by using groupby and count methods
df.groupby("Name", maintain_order = True).agg(pl.count())
```

```python
# Checking if there were anything recorded for the "Synonyms" column
df.groupby("Synonyms", maintain_order = True).agg(pl.count())
```

```python
# Interested in what types of "Max Phase" were recorded for the curated small molecules in ChEMBL database
df.groupby("Max Phase", maintain_order = True).agg(pl.count())
```

A quick groupby function showed that there were only 4,121 small molecules approved. Phase 3 recorded a total of 1,628 small molecules. The number was higher for phase 2, which had 2,346 small molecules, followed by 1,479 compounds in phase 1. There were, however, a massive total amount of 2,322,126 small molecules that had zero as phase number, which likely were either waiting to be assigned appropriate phase numbers or the compounds did not enter phase 1 at all.

Next the "Target" column was explored, all entries were shown as numbers. So what did all these numbers represent? I went through several places on ChEMBL database website. Eventually I've settled with an answer that was previously asked by someone, which was relevant or similar to what I wanted to ask (screenshots for ChEMBL cpd related targets). Each compound in the dataset had the number of different targets related to it shown in the "Target" column.

```python
# Here I tried to find out how many times each number of targets appeared in the dataset by
# using the simple groupby and count functions, keeping the dataframe in the original order
df.groupby("Targets", maintain_order = True).agg(pl.count())
```

```python
# Sorting the "Targets" column in a descending order
df.sort("Targets", reverse = True)
```

```python
# Then I tried to sort it similarly using "Targets" column with also the "ChEMBL ID" column, 
# showing the highest number of targets recorded appeared to be 994!
df.sort(
    [pl.col("Targets"), pl.col("ChEMBL ID") ** 2],
    reverse = [True, False],
)
```

```python
# This was confirmed via a filter function, which brought up CHEMBL1421, or also known as dasatinib
df.filter(pl.col("Targets") == "994")
```

```python
# Then I tested the filter function, so I searched for compounds with one target only, 
# zero violations of the rule-of-five and achieved a max phase of 4 which meant they were approved compounds
# This brought up a total of 92 compounds as shown below
df.filter((pl.col("Targets") == "1") & (pl.col("#RO5 Violations") == "0") & (pl.col("Max Phase") == 4))
```
